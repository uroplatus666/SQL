SELECT pc1.model AS max_model, pc3.model AS min_model, pc1.speed, pc1.ram
FROM PC pc1
INNER JOIN
(SELECT speed, ram
FROM PC
GROUP BY speed, ram
HAVING COUNT(*) = 2) pc2
ON pc1.speed = pc2.speed AND pc1.ram = pc2.ram
INNER JOIN PC pc3
ON pc2.speed = pc3.speed AND pc2.ram = pc3.ram AND pc3.model < pc1.model
